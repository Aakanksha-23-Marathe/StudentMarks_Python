```python
#importing required libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#importing dataset
df = pd.read_csv('StudentPerformance_Data_Analysis.csv')
```

```python
#analysing top 5 records of dataframe
df.head()
```

```python
#describe numerice columns
df.describe()
```

```python
#analysing basic informaiton of data
df.info()
```

```python
#find null values availability
df.isnull().sum()
```

```python
#drop unnamed column
df= df.drop('Unnamed: 0', axis = 1)
df.head()
```

```python
#change WklyStudyHours columns value 'o5-oct' to '5-10'
df.WklyStudyHours.unique()
df['WklyStudyHours'] = df['WklyStudyHours'].str.replace('05-Oct','5-10')
df.head()
```

```python
# Count plots for categorical variables
categorical_cols = ['Gender', 'EthnicGroup', 'ParentEduc', 'LunchType', 'TestPrep', 'ParentMaritalStatus', 'PracticeSport', 'IsFirstChild', 'TransportMeans', 'WklyStudyHours']
for col in categorical_cols:
    plt.figure(figsize=(8,5))
    ax = sns.countplot(data=df, x=col)
    ax.bar_label(ax.containers[0])
    plt.title(f'Distribution of {col}')
    plt.show()
```

From the above chart we have analysed:
1. the number of females in the data is more than the number of males.
2. Group C has max distribution whereas Group A has minimum
3. Max Student's parents have some college degree
4. Max student's have Standard lunchtype
5. Max Student's parents are married
6. Max Students practise sports regularly
7. Max Students prefer school_bus over private
8. Max Students study between 5 to 10 hrs weekly

```python
# Distribution Comparison - Violin Plots
categorical_cols = ['Gender', 'EthnicGroup', 'ParentEduc', 'LunchType', 'TestPrep', 'ParentMaritalStatus', 'PracticeSport', 'IsFirstChild', 'TransportMeans']
for col in categorical_cols:
    plt.figure(figsize=(12, 6))
    sns.violinplot(data=df, x=col, y='MathScore', inner='quartile')
    plt.title(f'Violin Plot of {col} vs. MathScore')
    plt.show()
```

```python
#analysing impact of parents education on students scores
gb = df.groupby('ParentEduc').agg({'MathScore':'mean','ReadingScore':'mean','WritingScore':'mean'})
print(gb)
```

```python
#analysing impact of parents education on students scores through graphs
plt.figure(figsize=(4,4))
sns.heatmap(gb, annot = True)
plt.title("Relationship between Parent's Education and Student's Score")
plt.show()
```

From the above chart, we have concluded that the education of the parents have a good impact on students score. 
Students having parents with masters degree have scored well compared to others.

```python
#Analysing impact of parent's marital status on student's score
pm = df.groupby('ParentMaritalStatus').agg({'MathScore':'mean','ReadingScore':'mean','WritingScore':'mean'})
print(pm)
```

```python
#analysing impact of parents marital status on students scores through graphs
plt.figure(figsize=(4,4))
sns.heatmap(pm, annot = True)
plt.title("Relationship between Parent's Marital Status and Student's Score")
plt.show()
```

From the above chart, we have concluded that the marital status of the parents have a negligible impact on students score.

```python
#Analysing ethnic groups
print(df['EthnicGroup'].unique())
```

```python
#distribution of Ethnic Groups
groupA = df.loc[(df['EthnicGroup']=="group A")].count()
groupB = df.loc[(df['EthnicGroup']=="group B")].count()
groupC = df.loc[(df['EthnicGroup']=="group C")].count()
groupD = df.loc[(df['EthnicGroup']=="group D")].count()
groupE = df.loc[(df['EthnicGroup']=="group E")].count()

mlist=[groupA["EthnicGroup"],groupB["EthnicGroup"],groupC["EthnicGroup"],groupD['EthnicGroup'],groupE["EthnicGroup"]]
l = ['group A','group B','group C','group D','group E']
plt.pie(mlist,labels = l,autopct = '%1.2f%%')
plt.title("Distribution of Ethnic Groups")
plt.show()
```

From above chart, we can see Group C has max distribution

```python
#Analysing Relationship between Wkly Study Hrs and Scores of student

# Scatter plot for MathScore vs. Weekly Study Hours
sns.scatterplot(data=df, x='WklyStudyHours', y='MathScore')
plt.title('MathScore vs. Weekly Study Hours')
plt.show()

# Scatter plot for WritingScore vs. Weekly Study Hours
sns.scatterplot(data=df, x='WklyStudyHours', y='WritingScore')
plt.title('WritingScore vs. Weekly Study Hours')
plt.show()
# Scatter plot for ReadingScore vs. Weekly Study Hours
sns.scatterplot(data=df, x='WklyStudyHours', y='ReadingScore')
plt.title('ReadingScore vs. Weekly Study Hours')
plt.show()

# Adjust layout to prevent overlapping
plt.tight_layout()

# Show the plot
plt.show()
```

From Above Chart, we can see there is no such impact of study hours on exam score
